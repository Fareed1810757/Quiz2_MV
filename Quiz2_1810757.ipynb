```python
def displayImage(image):
    if len(image.shape)==3:
        color_coverted = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        plt.imshow(color_coverted)
        plt.show()
        
    else:
        plt.imshow(image, cmap="gray")
        plt.show()
```

```python
img = cv2.imread(path+'lego1.PNG')
grayImage =cv2.cvtColor(img,cv2.COLOR_RGB2GRAY) #convert RGB image to GRAY
```

```python
displayImage(grayImage)
```

```python
blurred = cv2.GaussianBlur(grayImage, (7,7),25) #perform Gaussian Blur
edged = cv2.Canny(blurred, 10, 45)  #perform canny edge detection
```

```python
displayImage(edged)
```

```python
# define a (3, 3) structuring element
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3, 3)) #perform morphological transforms
```

```python
# apply the dilation operation to the edged image
dilate = cv2.dilate(edged, kernel, iterations=5) #perform dilation with 5 iterations
```

```python
(cnt,heirarchy)=cv2.findContours(dilate.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE) #retrieves contours from image
rgb=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
cv2.drawContours(rgb, cnt, -1, (0,255,0),2) #draw counter outlines

plt.imshow(rgb)
```
